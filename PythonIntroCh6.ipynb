## 6.2 De `for` Loop
In feite doet de `for`-lus iets voor elke waarde in een lijst. De manier waarop het is uiteengezet is een beetje verwarrend, maar verder is het erg basic. Hier is een voorbeeld ervan in code:

````python
# Voorbeeld 'for' loop
# Maak eerst een list om doorheen te "loopen":
nieuweList = [45, "Antwerpen", 90210, "We gaan een stukje fietsen", -67]

# maak de loop:
# Ga door nieuweList, en zet achtereenvolgens elk stukje informatie
# in de variabele waarde, en voer de lus uit 
for waarde in nieuweList:
    print(waarde)
````

Zoals je ziet, doorloopt de lus, wanneer hij wordt uitgevoerd, alle waarden in de list die na `in` wordt genoemd. Vervolgens worden ze in de variabele `waarde` gezet en door de lus uitgevoerd, waarbij elke keer dat de waarde iets anders is. Laten we het nog eens bekijken. We gaan een voetbalsupporter programmaatje maken: 

````python
# voetbalsupporter programma
woord = input("Voor wie ga je?")

for letter in woord:
    roep = "Geef me een " + letter + "!"
    print(roep)
    print(letter + "!")

print("Welk woord wordt hier gespeld")
print(woord + "!")
````

Je hebt zojuist een aantal dingen geleerd:
* Zoals je ziet, zijn strings (onthoud - strings zijn regels tekst) slechts lists met veel tekens.
* Het programma doorliep elk van de letters (of waarden) in woord en drukte ze op het scherm af.

En dat is alles wat betreft de `for` loop.

## 6.3 Een menu functie maken
Nu kunnen we kijken naar een toepassing van de `for` loop in programma's. Tot nu toe hebben we variabelen, lists, loops en functies geleerd. Dat is vrijwel alles wat we nodig hebben als basis om al wat te kunnen programmeren. We gaan dit toepassen in een functie om een keuzemenu te maken.

```Python
# DE MENU FUNCTIE
# Het programma vraagt om een string met alle menu-opties erin,
# en een string die een vraag stelt.
# zorg ervoor dat elk menu-item uniek is. 

def menu(list, vraag):
    for optie in list:
        print(1 + list.index(optie),end="")
        print (") " + optie)

    return eval(input(vraag)) - 1

# def menu(list, vraag): vertelt de functie om twee stukjes
# twee stukjes informatie te vragen:
# Een list met alle menu-items,
# en de vraag die het zal stellen als alle opties zijn weergegeven

# for optie in list: zegt zo ongeveer;
#'voor iedere optie in de list, doe het volgende:'

# print(list.index(optie) + 1 gebruikt de .index() functie om te vinden
# waar in de list de optie staat. print functie geeft het dan weer
# het telt 1 op om de getallen begrijpelijker te maken .

# print ") " + optie geeft een haakje weer en dan de optie.

# nadat de for-lus is voltooid, stelt eval(input (vraag) - 1) de vraag
# en geeft de waarde terug aan het hoofdprogramma (min 1, om het terug te zetten
# naar het getal dat de computer zal begrijpen).
```

## 6.4 Ons eerste 'spel'
Wat wordt ons eerste voorbeeldprogramma? Wat dacht je van een (heel) eenvoudig tekstavonturenspel? Lijkt me leuk! Het zal slechts één kamer van een huis omvatten en zal uiterst eenvoudig zijn. Er zullen vijf voorwerpen zijn, en een deur. In een van de vijf voorwerpen zit een sleutel om de deur te openen. Je moet de sleutel vinden en dan de deur openen. Ik zal eerst een gewone Nederlandse versie geven (pseudocode), en doe het vervolgens in Python: 

```
#Pseudocode van ons 'spel'

Vertel de computer over onze menufunctie

Geef een welkomstbericht weer met een beschrijving van de kamer.
We geven de speler zes voorwerpen om naar te kijken: bloempot, schilderij, \
  vaas, lampenkap, schoen en de deur 

Vertel de computer dat de deur op slot is
Vertel de computer waar de sleutel is

Geef een menu weer dat vertelt welke voorwerpen je kan 'gebruiken':
     Het geeft je de zes opties
     Het zal de vraag stellen "wat ga je bekijken?"

als de gebruiker wilde kijken naar:
     plant:
         Als de sleutel hier is, geef de speler de sleutel
         vertel anders dat de sleutel niet hier is
     schilderij:
         hetzelfde als hierboven
     enz.
     deur:
         Als de speler de sleutel heeft, laat hem dan de deur openen
         Zeg anders dat ze beter moeten zoeken 

Geef de speler een bericht voor het succesvol voltooien van het spel.
```
Hieruit kunnen we een echt programma schrijven. Ben je zover? Hier komt het: 

````python
# TEKSTAVONTURENSPEL

#De menu functie:
def menu(list, vraag):
    for optie in list:
        print(1 + list.index(optie),end="")
        print (") " + optie)

    return eval(input(vraag)) - 1

#Geef de computer wat basisinformatie over de kamer:
voorwerpen = ["bloempot","schilderij","vaas","lampenkap","schoen","deur"]

#De sleutel zit in de vaas (of nummer 2 in de bovenstaande list):
sleutellocatie = 2

#Je hebt de sleutel niet gevonden
sleutelgevonden = 0

loop = 1

#Geef een inleidende tekst:
print("Gisteravond ben je thuis in slaap gevallen.")

print("Nu merk je dat je opgesloten zit in een kamer. Je weet niet hoe")
print("je daar gekomen bent of hoe laat het is. In de kamer zie je")
print(len(voorwerpen), "voorwerpen:")
for x in voorwerpen:
    print(x)
print("")
print("De deur zit op slot. Zou er ergens een sleutel kunnen zijn?")
#Laat uw menu werken en het programma wordt uitgevoerd totdat je de sleutel vindt:
while loop == 1:
    keuze = menu(voorwerpen,"Welk voorwerp zou je willen bekijken?")
    if keuze == 0:
        if keuze == sleutellocatie:
            print("Je hebt een kleine sleutel in de bloempot gevonden.")

            print("")
            sleutelgevonden = 1
        else:
            print("Je hebt niets gevonden in de bloempot.")
            print("")
    elif keuze == 1:
        if keuze == sleutellocatie:
            print("Je hebt een kleine sleutel achter het schilderij gevonden.")
            print("")
            sleutelgevonden = 1
        else:
            print("Je hebt niets gevonden achter het schilderij.")
            print("")
    elif keuze == 2:
        if keuze == sleutellocatie:
            print("Je hebt een kleine sleutel gevonden in de vaas.")
            print("")
            sleutelgevonden = 1
        else:
            print("Je hebt niets gevonden in de vaas.")
            print("")
    elif keuze == 3:
        if keuze == sleutellocatie:
            print("Je hebt een kleine sleutel gevonden in de lampenkap.")
            print("")
            sleutelgevonden = 1
        else:
            print("Je hebt niets gevonden in de lampenkap.")
            print("")

    elif keuze == 4:
        if keuze == sleutellocatie:
            print("Je hebt een kleine sleutel gevonden in de schoen.")
            print("")
            sleutelgevonden = 1
        else:
            print("Je hebt niets gevonden in de schoen.")
            print("")
    elif keuze == 5:
        if sleutelgevonden == 1:
            loop = 0
            print("Je steekt de sleutel in het sleutelgat, draait hem om en hoort een klik.")
            print("")
        else:
            print("De deur is op slot, je moet een sleutel vinden.")
            print("")

# Onthoud dat een backslash de code 
# continueert op de volgende regel
print("Licht stroomt de kamer binnen als \
je de deur opent naar je vrijheid.")
````

Een heel eenvoudig, maar leuk spel. Laat je niet afschrikken door de hoeveelheid code. 53 van de regels zijn slechts de `if`-instructies, wat makkelijk is om te lezen als je eenmaal alle inspringingen begrijpt.
